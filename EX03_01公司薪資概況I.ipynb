取得 Name 和 Salary 二欄資料

```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}
df = pd.DataFrame(data)
df
```

取出索引值 5 到 8 共 4 筆資料

```python
avg_age = df['Age'].mean() #年齡平均值
avg_salary = df['Salary'].mean() #薪水平均值
df[(df['Age'] > avg_age) & (df['Salary'] > avg_salary)] #年齡、薪水是否大於平均值
```

請問公司人員的平均薪水、及最高、最低薪水、中位數？

```python
print(df['Salary'].mean(), df['Salary'].max(), df['Salary'].min(), df['Salary'].median()) #薪水平均數、最大值、最小值、中位數
```

公司共有幾個部門？

```python
df['Department'].unique() #公司有幾個部門
```

那個部門薪水最高？那個部門最年輕？

```python
print(df.groupby('Department')['Salary'].mean().idxmax(), df.groupby('Department')['Salary'].mean().max()) #把部門群組化 #部門薪水平均值 #idxmax最高平均數的「索引值」 #部門薪水平均值
```

那個部門最年輕？

```python
print(df.groupby('Department')['Age'].mean().idxmin(), df.groupby('Department')['Age'].mean().min()) #同上，但年紀最小平均值
```

年齡是否與薪資有關？

```python
corr_matrix = df[['Age', 'Salary']].corr() #年齡、薪水相關係數
corr_matrix
```
